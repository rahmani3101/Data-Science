Row Selection from dataframe

 Understanding `df.loc[]` and `df.iloc[]` for Data Selection

`df.loc[]` and `df.iloc[]` are two fundamental methods in Pandas for selecting data from a DataFrame. While both are used for selection, they operate based on different principles:

1.  **`df.loc[]` (Label-based indexing)**
    *   **Selects data by *explicit labels*** (names) of rows and columns.
    *   When slicing (e.g., `start:end`), *both the start and end labels are included* in the selection.
    *   It's ideal when you know the exact names of the rows or columns you want.

2.  **`df.iloc[]` (Integer-location based indexing)**
    *   **Selects data by *integer positions*** (0-based index) of rows and columns.
    *   Similar to standard Python slicing, when slicing (`start:end`), the `start` position is included, but the `end` position is *excluded*.
    *   It's useful when you want to select data based on its numerical position, regardless of its label.

Let's demonstrate these with the current DataFrame `df`.

```python
d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']), 'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
df
```

There are two ways of indexing Dataframes : Label based indexing and Boolean Indexing.

## Label Based Indexing

DataFrame.loc[] is an important method that is used for label based indexing with DataFrames.

```python
d = {'one' : pd.Series([5,6,7,8], index=['row1', 'row2', 'row3','row4']), 'two' : pd.Series([1, 2, 3, 4], index=['row1', 'row2', 'row3','row4']), 'three' : pd.Series([9, 10, 3, 1], index=['row1', 'row2', 'row3','row4'])}
df = pd.DataFrame(d)
df
```

```python
print(df.loc[['row2','row3']])
```

```python
print(df.loc[:,'two'])
```

```python
print(df['two'])
```

## Boolean Indexing

In Boolean indexing, we can select the subsets of data based on the actual values in the DataFrame rather than their row/column labels. Thus, we can use conditions on column names to filter data values.

```python
d = {'one' : pd.Series([5,6,7,8], index=['row1', 'row2', 'row3','row4']), 'two' : pd.Series([1, 2, 3, 4], index=['row1', 'row2', 'row3','row4']), 'three' : pd.Series([9, 10, 3, 1], index=['row1', 'row2', 'row3','row4'])}
df = pd.DataFrame(d)
df
```

```python
print(df.loc['row2'] > 3)
```

```python
print(df.loc[:,'two'] > 3)
```

## Joining, Merging and Concatenation of DataFrames

```python
dF1=pd.DataFrame([[1, 2, 3], [4, 5],[6]], columns=['C1', 'C2', 'C3'], index=['R1','R2', 'R3'])
print(dF1)
dF2=pd.DataFrame([[10, 20], [30], [40,50]], columns=['C2', 'C5'], index=['R4', 'R2','R5'])
print(dF2)
dF1=pd.concat([dF1, dF2])
print(dF1)
```

### Using `pd.merge()` method

`pd.merge()` is used to combine DataFrames based on common columns or indices. It's similar to SQL JOIN operations. You can merge on one or multiple conditions (keys).

```python
DF1 = pd.DataFrame({
    'id':[1,2,3],
    'Name': ['Ajay', 'Bijay', 'Sujoy'],
    'subject':['sub1','sub2','sub4']
})
DF2 = pd.DataFrame({
    'id':[1,2,3],
    'Name': ['Rakhi', 'Sakhi', 'Malini'],
    'subject':['sub1','sub5','sub4']
})

print("DF1:\n", DF1)
print("\nDF2:\n", DF2)
```

#### Joining with `merge()` on a single condition ('id')

```python
print ("Merged on 'id':")
print (pd.merge(DF1, DF2, on='id'))
```

#### Joining with `merge()` on two conditions ('id' and 'subject')

```python
print ("\nMerged on ['id', 'subject']:")
print (pd.merge(DF1, DF2, on=['id','subject']))
```

### Using `pd.concat()` method

The `pd.concat()` function performs concatenation operations along an axis, either stacking DataFrames row-wise (axis=0, default) or column-wise (axis=1).

```python
DF1 = pd.DataFrame({'id':[1,2,3], 'Name': ['Ajay', 'Bijay', 'Sujoy'], 'subject':['sub1','sub2','sub4']},index=[1,2,3])
DF2 = pd.DataFrame({'id':[1,2,3], 'Name': ['Rakhi', 'Sakhi', 'Malini'], 'subject':['sub1','sub5','sub4']},index=[1,2,3])

print("DF1 (for concat):\n", DF1)
print("\nDF2 (for concat):\n", DF2)
```

#### Joining with `concat()` with default axis (axis=0 - row-wise)

```python
print ("\nConcatenated (axis=0, default):")
print (pd.concat([DF1,DF2]))
```

#### Joining with `concat()` with axis=1 (column-wise)

```python
print ("\nConcatenated (axis=1):")
print (pd.concat([DF1,DF2],axis=1))
```
