```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
x = np.array([1,2,3])
y = np.array([4,5,6])

xx , yy = np.meshgrid(x,y)
print("Co ordinate matrix xx:")
print(xx)
print("\n Co ordinate matrix of yy :")
print(yy)

z = xx**2 + yy**2
print("\n Function value z = x^2 + y^2 : ")
print(z)

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot_surface(xx,yy,z , cmap = 'viridis')

ax.set_xlabel('X- Axis')
ax.set_ylabel('Y- Axis')
ax.set_zlabel('Z- Axis')
ax.set_title('Surface plot of f(x,y) = x^2 + y^2 ')
plt.show()      
```

```python
import plotly.graph_objects as go

x= np.array([1,2,3])
y = np.array([4,5,6])
xx , yy = np.meshgrid(x,y)
z = xx**2 + yy**2
fig = go.Figure(data = [go.Surface(x=xx,y=yy,z=z)])
fig.update_layout(title = 'Surafce plot of f(x,y) = x*2 + y^2',
                  scene = dict(xaxis_title='X-Axis',
                               yaxis_title = 'Y-Axis',
                               zaxis_title = 'Z-Axis'))

fig.show()
```
